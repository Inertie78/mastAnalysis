# üå¨Ô∏è √âquilibre d‚Äôun m√¢t hauban√© √† deux niveaux

## ‚öôÔ∏è 1. Donn√©es du probl√®me

Le m√¢t est une poutre triangul√©e en aluminium **tube √ò50√ó2 mm**, de hauteur **h = 16 m**, hauban√© √† **deux niveaux** :

| Param√®tre | Symbole |
|:-----------|:---------:|
| masse de la lampe | $m_{lampe}$ |
| masse du m√¢t | $m_{mat}$ |
| Hauteur du m√¢t | $ha_{mat}$ |
| Diam√®tre tube | D |
| Coefficient de tra√Æn√©e du tube | $C_d$ |
| Distance d‚Äôancrage | $I_{anc}$ |
| Surface de la lampe | $A_{lampe}$ |
| Pression du vent | $p_{vent}$ |

### üí® R√©partition du vent

Le vent agit sur :
- **la lampe** (surface frontale),
- **le m√¢t triangul√©** (surface projet√©e √©quivalente).

Force totale due au vent :

$$
F_{vent} = F_{lampe} + F_{m√¢t}
$$

avec :

$$
F_{lampe} = p_{vent} \times A_{lampe}
$$

et

$$
F_{m√¢t} = p_{vent} \times D \times h \times C_d
$$

üí° Pour un m√¢t treillis, on peut appliquer un **facteur de porosit√©** (‚âà0.4) pour ajuster la charge effective.

---

## üß© 2. Angles et g√©om√©trie

Les haubans sont dispos√©s √† 120¬∞ dans le plan horizontal, √† deux niveaux :

$$
\tan(\alpha_1) = \frac{h}{I_{anc}}, \quad 
\tan(\alpha_2) = \frac{h/2}{I_{anc}}
$$

soit :

$$
\alpha_1 = 63.4^\circ, \quad \alpha_2 = 45^\circ
$$

---

## ü™ú 3. R√©partition de la charge du vent

Le profil du vent augmente avec la hauteur.  
En supposant une loi de puissance $p(z) \propto z^\alpha$ avec $\alpha = 0.2$ :

$$
\frac{F_{haut}}{F_{total}} = 1 - \left(\frac{1}{2}\right)^{1+\alpha} \approx 0.64
$$

On adopte donc une r√©partition **64 % ‚Äì 36 %** entre les deux niveaux :

$$
F_{vent,haut} = 0.64 \, F_{vent}, \quad F_{vent,bas} = 0.36 \, F_{vent}
$$

---

## üßÆ 4. √âquilibre des forces et des moments

### a) √âquilibre horizontal (axe du vent) :

$$
R_x + \cos(\alpha_1)(T_{1,1} + T_{a,1}) + \cos(\alpha_2)(T_{1,2} + T_{a,2}) = F_{vent}
$$

### b) √âquilibre lat√©ral (perpendiculaire au vent) :

Les haubans non sollicit√©s dans le plan du vent g√©n√®rent une composante lat√©rale √©quilibr√©e par la r√©action au sol $R_y$ :

$$
R_y = -\frac{\sqrt{3}}{2} \left[ \cos(\alpha_1) (T_{a,1}) + \cos(\alpha_2) (T_{a,2}) \right]
$$

Cette r√©action assure la **stabilit√© tridimensionnelle** du m√¢t.

### c) √âquilibre des moments autour du pied :

La r√©sultante du vent agit √† une hauteur de **0.6 h** :

$$
F_{vent} \times 0.6h = \cos(\alpha_1)(T_{1,1} + T_{a,1}) \, h + \cos(\alpha_2)(T_{1,2} + T_{a,2}) \, \frac{h}{2}
$$

### d) √âquilibre vertical :

$$
R_z = F_{ec} + P_{m√¢t} + 3 (T_{1,1}\sin(\alpha_1) + T_{1,2}\sin(\alpha_2))
$$

---

## üß∑ 5. Hypoth√®ses de r√©partition

Dans chaque niveau :

$$
T_{1,i} = 2 \, T_{a,i}
$$

ce qui correspond √† un hauban **dans l‚Äôaxe du vent** deux fois plus charg√© que les deux lat√©raux.

Ainsi :

$$
\begin{cases}
T_{1,1} = \dfrac{4 F_{vent,haut}}{9 \cos(\alpha_1)} \\[8pt]
T_{a,1} = \dfrac{2 F_{vent,haut}}{9 \cos(\alpha_1)} \\[8pt]
T_{1,2} = \dfrac{2 F_{vent,bas}}{9 \cos(\alpha_2)} \\[8pt]
T_{a,2} = \dfrac{F_{vent,bas}}{9 \cos(\alpha_2)} \\[8pt]
R_x = -\dfrac{1}{2} \cos(\alpha_2)(T_{1,2} + T_{a,2}) \\[8pt]
R_y = -\dfrac{\sqrt{3}}{2} \left[\cos(\alpha_1)T_{a,1} + \cos(\alpha_2)T_{a,2}\right]
\end{cases}
$$

---

## üìä 6. Interpr√©tation physique

- Les **haubans sup√©rieurs** reprennent environ **64 %** de la charge du vent.  
- Les **haubans inf√©rieurs** stabilisent le m√¢t (‚âà36 % du vent).  
- Les **c√¢bles dans l‚Äôaxe du vent** sont les plus tendus.  
- La **r√©action horizontale** $R_x$ peut √™tre nulle si le pied est **articul√©**.  
- La **r√©action lat√©rale** $R_y$ √©quilibre les efforts hors plan.  
- La **r√©action verticale** $R_z$ inclut le poids du m√¢t et la lampe.

---

## üß† 7. Points √† v√©rifier ou affiner

### ‚öôÔ∏è R√©partition du vent (0.64 ‚Äì 0.36)
Issue d‚Äôun profil de vent r√©aliste $p(z) \propto z^{0.2}$.  
La diff√©rence avec 70‚Äì30 est n√©gligeable pour un dimensionnement pr√©liminaire.

### ‚öôÔ∏è Moment d‚Äô√©quilibre
Le moment de chaque c√¢ble s‚Äô√©crit :
$$
M_i = T_i \cos(\alpha_i) \, z_i
$$
et la r√©sultante du vent agit √† **0.6 h** (centre de pouss√©e).

### ‚öôÔ∏è R√©actions horizontales et lat√©rales
- $R_x$ : r√©action dans le plan du vent, transmise si le pied est encastr√©.  
- $R_y$ : r√©action lat√©rale, √©quilibre les efforts transversaux dans le plan perpendiculaire au vent.

### ‚öôÔ∏è Effort vertical $R_z$
Inclut le poids du m√¢t et la pr√©contrainte des haubans :
$$
R_z = P_{m√¢t} + P_{lampe} + 3(T_{1,1}\sin(\alpha_1) + T_{1,2}\sin(\alpha_2))
$$

### ‚öôÔ∏è V√©rification dimensionnelle
Lien avec la pression dynamique du vent :
$$
F_{vent} = \tfrac{1}{2} \rho C_d A V^2
$$
avec $C_d ‚âà 1.2$ pour un tube cylindrique, $\rho ‚âà 1.225 \, \mathrm{kg/m^3}$.

# ‚öñÔ∏è 9. D√©composition 3D des efforts dans les haubans

## üéØ Objectif
Chaque hauban est inclin√© d‚Äôun angle $\alpha_i$ avec l‚Äôhorizontale et orient√© √† **120¬∞** les uns des autres dans le plan horizontal.  
On veut exprimer les **composantes cart√©siennes $(X, Y, Z)$** des efforts de tension $T_{1,i}$ (dans le plan du vent) et $T_{a,i}$ (les deux autres).

---

## üß≠ 9.1 Orientation des haubans

On d√©finit l‚Äôaxe :
- $X$ : **axe du vent**,  
- $Y$ : **perpendiculaire au vent**,  
- $Z$ : **vertical (positif vers le haut)**.

Les trois haubans de chaque niveau $i$ sont orient√©s √† :
$$
\theta_{1,i} = 0^\circ, \quad
\theta_{2,i} = 120^\circ, \quad
\theta_{3,i} = 240^\circ
$$

---

## üìê 9.2 Composantes d‚Äôun hauban

Pour un hauban de tension $T_i$ inclin√© d‚Äôun angle $\alpha_i$ :

$$
\begin{cases}
T_{x} = T_i \cos(\alpha_i)\cos(\theta_i) \\
T_{y} = T_i \cos(\alpha_i)\sin(\theta_i) \\
T_{z} = -T_i \sin(\alpha_i)
\end{cases}
$$

> üî∏ Le signe ¬´ ‚Äì ¬ª en $T_z$ traduit que le hauban tire **vers le bas** le sommet du m√¢t.

---

## üßÆ 9.3 Application par niveau

### Niveau sup√©rieur ($i=1$, $\alpha_1$)

| Hauban | Tension | $\theta$ (¬∞) | $T_x$ | $T_y$ | $T_z$ |
|:-------|:--------:|:-------------:|:------:|:------:|:------:|
| Hauban avant (vent) | $T_{1,1}$ | 0 | $T_{1,1}\cos\alpha_1$ | 0 | $-T_{1,1}\sin\alpha_1$ |
| Hauban lat√©ral G | $T_{a,1}$ | 120 | $-\tfrac{1}{2}T_{a,1}\cos\alpha_1$ | $\tfrac{\sqrt{3}}{2}T_{a,1}\cos\alpha_1$ | $-T_{a,1}\sin\alpha_1$ |
| Hauban lat√©ral D | $T_{a,1}$ | 240 | $-\tfrac{1}{2}T_{a,1}\cos\alpha_1$ | $-\tfrac{\sqrt{3}}{2}T_{a,1}\cos\alpha_1$ | $-T_{a,1}\sin\alpha_1$ |

---

### Niveau inf√©rieur ($i=2$, $\alpha_2$)

| Hauban | Tension | $\theta$ (¬∞) | $T_x$ | $T_y$ | $T_z$ |
|:-------|:--------:|:-------------:|:------:|:------:|:------:|
| Hauban avant (vent) | $T_{1,2}$ | 0 | $T_{1,2}\cos\alpha_2$ | 0 | $-T_{1,2}\sin\alpha_2$ |
| Hauban lat√©ral G | $T_{a,2}$ | 120 | $-\tfrac{1}{2}T_{a,2}\cos\alpha_2$ | $\tfrac{\sqrt{3}}{2}T_{a,2}\cos\alpha_2$ | $-T_{a,2}\sin\alpha_2$ |
| Hauban lat√©ral D | $T_{a,2}$ | 240 | $-\tfrac{1}{2}T_{a,2}\cos\alpha_2$ | $-\tfrac{\sqrt{3}}{2}T_{a,2}\cos\alpha_2$ | $-T_{a,2}\sin\alpha_2$ |

---

## üìä 9.4 Somme vectorielle des efforts par niveau

En sommant les trois haubans d‚Äôun m√™me niveau :

$$
\boxed{
\begin{aligned}
\sum T_x &= \cos(\alpha_i)(T_{1,i} - T_{a,i}) \\
\sum T_y &= 0 \\
\sum T_z &= -\sin(\alpha_i)(T_{1,i} + 2T_{a,i})
\end{aligned}
}
$$

Ces r√©sultats confirment :
- **Effort horizontal** dans le plan du vent ‚Üí proportionnel √† $(T_{1,i} - T_{a,i})$  
- **Aucune r√©sultante lat√©rale ($Y$)** (sym√©trie)  
- **Effort vertical** descendant d√ª √† la tension des haubans.

---

## ‚öñÔ∏è 9.5 √âquilibre global 3D du m√¢t

En notant la somme sur les deux niveaux :

$$
\boxed{
\begin{aligned}
R_x + \sum_i \sum T_{x,i} &= F_{vent} \\
R_y + \sum_i \sum T_{y,i} &= 0 \\
R_z + \sum_i \sum T_{z,i} + P_{m√¢t} + P_{lampe} &= 0
\end{aligned}
}
$$

o√π chaque $\sum T_{x,i}$, $\sum T_{y,i}$, $\sum T_{z,i}$ est d√©fini comme ci-dessus.

---

## üß† 9.6 Interpr√©tation

- Les haubans **dans le plan du vent** apportent la r√©sistance principale ($T_x$ positif vers l‚Äôarri√®re).  
- Les haubans **lat√©raux** assurent la **stabilit√© transversale** ($T_y$ √©quilibr√©s).  
- Tous les haubans contribuent √† la **compression verticale** du m√¢t ($T_z$ n√©gatif).  



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ================================================================
# === 1Ô∏è‚É£ Donn√©es g√©n√©rales
# ================================================================

# --- Donn√©es physiques ---
ha_mat = 16.0       # hauteur du m√¢t (m)
Ianc = 6.0          # distance d‚Äôancrage (m)
Cd = 1.2            # coefficient de tra√Æn√©e du tube
dia_ext = 0.05      # diam√®tre ext√©rieur du tube (m)
dia_int = 0.046     # diam√®tre int√©rieur du tube (m)
ent_tube = 0.29    # entraxe entre tubes (m)
A_lampe = 2.0       # surface de la lampe (m¬≤)
m_lampe = 400       # masse de lampe kg
m_mat = 25 * 4      # masse du mat kg


# Constante
g = 9.81            # gravit√© terrestre
E = 70e9            # [Pa] (module d‚Äô√©lasticit√© de l‚Äôaluminium)
p_vent = 2700       # pression du vent (N/m¬≤)

# Poids propre du m√¢t (approximatif)
P_mat = m_mat * g  # N

# Force horizontale sur la lampe (vent sur lampe)
P_lampe = m_lampe * g

# charge verticale dans le mat du au mat et √† la lampe (N)
Fec = P_mat + P_lampe       

print(f"Poids du mat: {P_mat} N")
print(f"Poids de la lampe: {P_lampe} N")
print(f"Poids total {Fec} N")
```

```python
# ================================================================
# === 2Ô∏è‚É£ Forces du vent
# ================================================================
Fvent_lampe = p_vent * A_lampe
Fvent_mat = p_vent * dia_ext * ha_mat * Cd
Fvent_total = Fvent_lampe + Fvent_mat

# Moment du vent autour du pied du m√¢t
M_vent = Fvent_lampe * ha_mat + Fvent_mat * 0.6 * ha_mat

# R√©partition du vent entre les deux niveaux
Fhaut = (2 * M_vent / ha_mat - Fvent_total)
Fbas  = Fvent_total - Fhaut
Fhaut_ratio = Fhaut / Fvent_total
Fbas_ratio  = Fbas / Fvent_total


# ================================================================
# === 3Ô∏è‚É£ Angles des haubans
# ================================================================
alpha1 = sp.atan(ha_mat / Ianc)
alpha2 = sp.atan((ha_mat / 2) / Ianc)


# ================================================================
# === 4Ô∏è‚É£ Tensions dans les haubans
# Hypoth√®se : T1,i = 2 * Ta,i
# ================================================================
Ta1 = Fhaut / (3 * sp.cos(alpha1))
T11 = 2 * Ta1
Ta2 = Fbas / (3 * sp.cos(alpha2))
T12 = 2 * Ta2

Tsup = float(T11)
Tinf = float(T12)


# ================================================================
# === 5Ô∏è‚É£ R√©actions au pied du m√¢t
# ================================================================
Rx = -0.5 * sp.cos(alpha2) * (T12 + Ta2)
Ry = -sp.sqrt(3)/2 * (sp.cos(alpha1)*Ta1 + sp.cos(alpha2)*Ta2)
Rz = Fec + 3 * (T11 * sp.sin(alpha1) + T12 * sp.sin(alpha2))


# ================================================================
# === 6Ô∏è‚É£ D√©composition 3D des haubans
# ================================================================
def hauban_components(t, alpha, theta_deg):
    """Retourne (Tx, Ty, Tz) pour une tension T, un angle alpha et une orientation Œ∏."""
    theta = np.deg2rad(theta_deg)
    tx = t * np.cos(alpha) * np.cos(theta)
    ty = t * np.cos(alpha) * np.sin(theta)
    tz = -t * np.sin(alpha)
    return np.array([tx, ty, tz])

# Haubans sup√©rieurs (niveau 1)
H11 = hauban_components(float(T11), float(alpha1), 0)
HA1g = hauban_components(float(Ta1), float(alpha1), 120)
HA1d = hauban_components(float(Ta1), float(alpha1), 240)

# Haubans inf√©rieurs (niveau 2)
H12 = hauban_components(float(T12), float(alpha2), 0)
HA2g = hauban_components(float(Ta2), float(alpha2), 120)
HA2d = hauban_components(float(Ta2), float(alpha2), 240)

# --- R√©sultantes par niveau ---
Sum_Tx1 = H11[0] + HA1g[0] + HA1d[0]
Sum_Ty1 = H11[1] + HA1g[1] + HA1d[1]
Sum_Tz1 = H11[2] + HA1g[2] + HA1d[2]

Sum_Tx2 = H12[0] + HA2g[0] + HA2d[0]
Sum_Ty2 = H12[1] + HA2g[1] + HA2d[1]
Sum_Tz2 = H12[2] + HA2g[2] + HA2d[2]

# --- Affichage principal ---
print("=== R√âSULTATS MISE √Ä JOUR ===\n")
vals = {
    "Œ±1 (¬∞)"              : float(sp.deg(alpha1)),
    "Œ±2 (¬∞)"              : float(sp.deg(alpha2)),
    "Fvent_total (N)"     : float(Fvent_total),
    "Fhaut (N)"           : float(Fhaut),
    "Fbas (N)"            : float(Fbas),
    "Fhaut_ratio (%)"     : float(Fhaut_ratio * 100),
    "Fbas_ratio (%)"      : float(Fbas_ratio * 100),
    "T11 (N)"             : float(T11),
    "Ta1 (N)"             : float(Ta1),
    "T12 (N)"             : float(T12),
    "Ta2 (N)"             : float(Ta2),
    "Rx (N)"              : float(Rx),
    "Ry (N)"              : float(Ry),
    "Rz (N)"              : float(Rz),
}
for k, v in vals.items():
    print(f"{k:<25s} : {v:10.2f}")
```

```python
# ================================================================
# === 10Ô∏è‚É£ Charge critique de flambage (Euler, haubans pris en compte)
# ================================================================

Ic = (np.pi / 64) * ((dia_ext**4) - (dia_int**4))
Atube = (np.pi / 4) * ((dia_ext**2) - (dia_int**2))
Itotal = 2 * Ic + 2 * (Ic + (Atube * ((ent_tube/2)**2)))

# --- Longueurs et coefficients de flambement ---
L1 = ha_mat / 2     # partie inf√©rieure (bas-haubans)
L2 = ha_mat / 2     # partie sup√©rieure (haut-haubans)
K1 = 0.7            # encastr√© - articul√©
K2 = 1.0            # articul√© - articul√©

# --- Charges critiques partielles ---
Fcr1 = (np.pi**2 * E * Itotal) / ((K1 * L1)**2)
Fcr2 = (np.pi**2 * E * Itotal) / ((K2 * L2)**2)

# --- Charge critique √©quivalente (tron√ßons en s√©rie) ---
Fcr_eq = (Fcr1 * Fcr2) / (Fcr1 + Fcr2)

# --- Effort r√©el de compression dans le m√¢t ---
effort_compression_montant = float(Rz)
Csecu = Fcr_eq / effort_compression_montant

# --- Affichage console ---
print("\n=== üî© Flambage d‚ÄôEuler corrig√© (avec haubans) ===")
print(f"Module E : {E:.2e} Pa")
print(f"Moment d‚Äôinertie total : {Itotal:.4e} m‚Å¥")
print(f"Fcr bas (encastr√©-articul√©) : {Fcr1:,.2f} N")
print(f"Fcr haut (articul√©-articul√©): {Fcr2:,.2f} N")
print(f"Fcr √©quivalent global       : {Fcr_eq:,.2f} N")
print(f"Effort compression r√©el      : {effort_compression_montant:,.2f} N")
print(f"Coefficient de s√©curit√©      : {Csecu:.2f}")
```
